```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
df = pd.read_csv('nilai_siswa.csv', sep=';')
df.info()
```

```python
df.head()
```

```python
df.describe()
```

```python
print("--- Statistik Nilai Keseluruhan ---")
print("Rata-rata (Mean) :", df['Nilai'].mean())
print("Nilai Tengah (Median):", df['Nilai'].median())
print("Modus (Nilai Sering Muncul):", df['Nilai'].mode()[0])
```

```python
nilai_produktif = df[df['Matpel'] == 'Produktif']
print("--- Data Nilai Siswa Mapel Produktif ---")
print(nilai_produktif)
```

```python
summary_nilai = df.groupby('Matpel')['Nilai'].agg(['max', 'min', 'mean'])
print(summary_nilai)
```

```python
rata_rata = df.groupby('Matpel')['Nilai'].mean()
plt.figure(figsize=(8, 5))
rata_rata.plot(kind='bar')
plt.title('Rata-Rata Nilai Siswa per Mata Pelajaran')
plt.xlabel('Mata Pelajaran')
plt.ylabel('Skor Rata-Rata')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='Matpel', y='Nilai', data=df)
plt.title('Distribusi Sebaran Nilai Siswa')
plt.show()
```

**Jawaban:** Berdasarkan data hasil olahan, mapel **Produktif** memiliki rata-rata tertinggi.

### 2. Mapel mana yang memiliki nilai terendah?
**Jawaban:** Mapel **Matematika** memiliki nilai minimum terendah.
